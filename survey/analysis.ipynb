```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
responses = pd.read_csv("responses.csv")
answers = pd.read_csv("answers.csv")

print(responses.head())
print(answers.head())
```

```python
columns = [f"Column {x}" for x in range(3, 24)]
label = ["Normal", "Unusual"]
color_normal = ["green", "red"]
color_unusual = ["red", "green"]

age_groups = ["Under 18", "18-34", "35-54", "Over 55"]
backgrounds = ["Little to no background", 
               "General interest or basic knowledge of concepts", 
               "Completed coursework, training, or certification",
               "Academic researcher or student specializing in a cybersecurity-related field",
               "Professional experience"]
```

```python
answers_col = answers['Answer']

for bg in backgrounds:
    df = responses[responses['background'].str.contains(bg)]
    fig, axes = plt.subplots(1, len(columns), figsize=(6 * len(columns), 6))
    
    i = 0
    for ax, col in zip(axes, columns):
        if (answers_col[i] == 'normal'):
            color = color_normal
        else:
            color = color_unusual
        i += 1
        df[col].value_counts().plot.pie(
            ax=ax,
            autopct='%1.1f%%',
            title=f"{col}",
            colors=color
        )
        ax.set_ylabel("")

    plt.tight_layout()
    print(f"{bg}")
    plt.show()
```
